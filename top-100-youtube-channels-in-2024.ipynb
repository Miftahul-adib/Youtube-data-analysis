# Dataset Preparation
    




```python
missing_values = df[df.isnull().any(axis=1)]
```

```python
df = df.dropna(subset=['Views','Uploads','Subscribers'])

```

```python
def convert_to_numeric(value):
    if isinstance(value, str):
        # Remove commas from the number
        value = value.replace(',', '')
        
        # Handle 'M' for millions in 'Subscribers'
        if 'M' in value:
            value = float(value.replace('M', '').strip()) * 1e6  # Convert 'M' to millions
        else:
            # Convert to float if it's just a regular number
            value = float(value)
    return value

# Apply the function to convert the columns
df['Subscribers'] = df['Subscribers'].apply(convert_to_numeric)
df['Views'] = df['Views'].apply(convert_to_numeric)

# Ensure the 'Uploads' column is of float type
df['Uploads'] = df['Uploads'].apply(convert_to_numeric)
# Convert 'Views' column to smaller, human-readable numbers (e.g., in millions)

```

```python
df['Views'] = (df['Views']).round().astype(int)
```

```python
# Update discrepancies with real numbers (without multiplication)
updates = {
    "PewDiePie": {"Country": "SE"},
    "EminemMusic": {"Country": "US"},
    "A4": {"Country": "BY"},
    "WWE": {"Country":"US"}
}

# Apply updates
for username, changes in updates.items():
    for column, new_value in changes.items():
        df.loc[df["Username"] == username, column] = new_value

```

```python
country_mapping = {
    'IN': 'India', 
    'US': 'United States', 
    'MX': 'Mexico', 
    'BR': 'Brazil', 
    'CA': 'Canada', 
    'KR': 'South Korea', 
    'PK': 'Pakistan', 
    'GB': 'United Kingdom', 
    'JP': 'Japan', 
    'AR': 'Argentina', 
    'IT': 'Italy', 
    'RU': 'Russia', 
    'CL': 'Chile', 
    'ES': 'Spain', 
    'PH': 'Philippines', 
    'TR': 'Turkey', 
    'SV': 'El Salvador', 
    'PR': 'Puerto Rico', 
    'AE': 'United Arab Emirates', 
    'ID': 'Indonesia', 
    'NO': 'Norway', 
    'CO': 'Colombia', 
    'PT': 'Portugal',
    'BY': 'Belarus',
    'SE': 'Sweden'
}

# Replace country codes with full country names
df['Country'] = df['Country'].map(country_mapping).fillna(df['Country'])
```

### Distribution of Views by Country (Top 5 by Total Channels in top 100 )

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter

# Function to format y-axis values
def format_y_axis(value, _):
    if value >= 1e6:
        return f'{value/1e6:.1f}M'  # Convert to millions
    elif value >= 1e3:
        return f'{value/1e3:.1f}K'  # Convert to thousands
    return f'{int(value)}'  # Show as is for smaller numbers

# Get the top 5 countries with the most YouTube channels
top_countries_by_channels = (
    df['Country']
    .value_counts()
    .nlargest(5)
    .index
)

# Filter the original DataFrame for these top 5 countries
df_top_countries = df[df['Country'].isin(top_countries_by_channels)].copy()

# Sorting countries by median views for better visualization
sorted_countries = (
    df_top_countries.groupby('Country')['Views']
    .median()
    .sort_values(ascending=False)
    .index
)
df_top_countries['Country'] = pd.Categorical(
    df_top_countries['Country'], 
    categories=sorted_countries, 
    ordered=True
)

# Setting the figure size
plt.figure(figsize=(12, 8))

# Creating the boxplot
sns.boxplot(
    x='Country', 
    y='Views', 
    data=df_top_countries, 
    palette='coolwarm', 
    showfliers=True  # Hides outliers for a cleaner view
)

# Adding customizations
plt.title('Distribution of Views by Country (Top 5 by Number of Channels)', fontsize=16, weight='bold')
plt.xlabel('Country', fontsize=14)
plt.ylabel('Views', fontsize=14)
plt.xticks(rotation=90, ha='right', fontsize=12)
plt.yticks(fontsize=12)

# Format the y-axis
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_y_axis))

# Tight layout for better spacing
plt.tight_layout()

# Displaying the plot
plt.show()
```



### **✨Insights**   

---

### 1) The range of views for YouTube channels from India is smaller than that of the United States, reflecting the greater diversity of U.S. viewers. 

### 2)U.S. channels benefit from a global audience and broader international appeal.

### 3)In contrast, Indian channels, despite their rapid growth, often focus on regional content, limiting their reach. However, some Indian channels have become notable outliers, achieving significant global popularity and expanding their audience beyond regional boundaries.
